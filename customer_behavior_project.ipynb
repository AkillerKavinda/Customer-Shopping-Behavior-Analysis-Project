```python
# Checking for null values

df[df['Review Rating'].isna()]
```

```python
# Imputing missing values of the Review Rating column with the median of their category 

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns to snake case

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# Feature engineering age groups

labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']

df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df.head()
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Feature engineering purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head()
```

```python
df[['discount_applied', 'promo_code_used']]
```

```python
# Checking if discount_applied and promo_code_used has the same values

(df['discount_applied']==df['promo_code_used']).all()
```

```python
# Dropping promo_code_used because it contains redundent information

df.drop(columns=['promo_code_used'], axis = 1, inplace=True)
```
